```python
def powerModInt(a,k,n):
    b = 1
    if k == 0:
        return b
    A = a
    if 1 & k: # If the least significant bit is 1, a^1 = a
        b = a
    k = k >> 1
    while k:
        A = (A**2) % n
        if 1 & k:
            b = (b * A) % n
        k = k >> 1
        
    return b

print(powerModInt(412545236899747100000,3,743330222539755158153))
```
