```python
#Importamos libreria de matemáticas

from collections import Counter
from fractions import Fraction
from decimal import Decimal

import operator
import numpy as np
```

```python
#Ejercicio 1:  calcula el mensaje

alfabeto = "ABCDEFGHIJKLMNÑOPQRSTUVWXYZ ."
cod = Fraction('0.1613657607216723798346110583')
long_msj = 19;

tam = len(alfabeto)
contador = Counter(alfabeto)

def metodoDecodificacion(long_msj,cod,contador):
    aux=0
    lista = [0]
    listaLetras = []
    freq_rel_acum = 0
    msg=""
    
    #Rellenamos la lista con los distitnos intervalos
    for clave , valor in contador.items():
            freq_rel_acum += Fraction(valor,tam)
            lista.append(freq_rel_acum)

    #print(lista)

    while(aux<long_msj):
        freq_rel_acum=0
        for clave , valor in contador.items():
            freq_rel_acum += Fraction(valor,tam)
            if(freq_rel_acum >= cod):
                caracter = clave
                msg += caracter
                break

        limitInf = lista[lista.index(freq_rel_acum)-1]
        print(" ")
        print("Letra: ",caracter)
        print("Limite Inferior: ",limitInf)
        print("Limite Superior: ",freq_rel_acum)
        
        cod = Fraction((cod-limitInf),(freq_rel_acum-limitInf))
        aux +=1

    print("\nMensaje:",msg)
    
metodoDecodificacion(long_msj,cod,contador)
```

```python
#Ejercicio 2:  calcula el mensaje

texto = "Existe una cosa muy misteriosa, pero muy cotidiana. Todo el mundo participa de ella,  todo el mundo la conoce, pero muy pocos se paran a pensar en ella.  Casi todos se limitan a tomarla como viene, sin hacer preguntas.  Esta cosa es el tiempo.  (Momo, Michael Ende)"
cod = Fraction('0.247276109705412160222')
long_msj = 17;

tam = len(texto)
contador = Counter(texto)

metodoDecodificacion(long_msj,cod,contador)

print(" ")

def metodoEntropia(tam):
    entropia = 0
    for clave , valor in contador.items():
        prob = valor/tam
        entropia += prob*np.log2(1/prob)
    print("La entropía de la fuente de texto es: ","{0:.5f}".format(entropia)) 

metodoEntropia(tam)
```
