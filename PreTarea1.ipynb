```python
#Ejercicio 1:  calcula el mensaje

#Importamos libreria de matemáticas

from collections import Counter
from fractions import Fraction

import operator
import numpy as np

alfabeto = "ABBCCCDEEFFFGHHIIIJKKLLLMNNÑÑÑOPPQQQRSSTTTUVVWWWXYYZZZ"
#long_msj = 6
#cod = Fraction('0.0187853247') #BALADA
#cod = Fraction('0.0560128422') #CAMADA
#cod = Fraction('0.685462733') #SALIDA

#long_msj = 8;
#cod = Fraction('0.420238079327') #LONGITUD

#long_msj = 10;
#cod = Fraction('0.4590768139525717') #MURCIELAGO

#long_msj = 12;
#cod = Fraction('0.446925244867976299') #MEDITERRANEO

long_msj = 12;
cod = Fraction('0.446925244867976299') #

tam = len(alfabeto)
contador = Counter(alfabeto)

def metodoDecodificacion(long_msj,cod):
    aux=0
    lista = [0]
    freq_rel_acum = 0
    
    #Rellenamos la lista con los distitnos intervalos
    for clave , valor in contador.items():
            freq_rel_acum += Fraction(valor,tam)
            lista.append(freq_rel_acum)

    print(lista)

    while(aux<long_msj):
        freq_rel_acum=0
        for clave , valor in contador.items():
            freq_rel_acum += Fraction(valor,tam)
            if(freq_rel_acum >= cod):
                caracter = clave
                break

        limitInf = lista[lista.index(freq_rel_acum)-1]
        print(" ")
        print("Letra: ",caracter)
        print("Limite Inferior: ",limitInf)
        print("Limite Superior: ",freq_rel_acum)

        cod = Fraction((cod-limitInf),(freq_rel_acum-limitInf))
        aux +=1
 
metodoDecodificacion(long_msj,cod)
```
